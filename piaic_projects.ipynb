```python
import random

def computer_guesses_number():
    """Function where the computer tries to guess the user's number"""
    print("\nWelcome to the Guess the Number Game - Computer Edition!")
    print("Think of a number between 1 and 100, and I'll try to guess it!")
    print("After each guess, tell me if I'm too (H)igh, too (L)ow, or (C)orrect!\n")

    # Initialize variables
    low = 1
    high = 100
    attempts = 0
    guessed_correctly = False

    # Game loop
    while not guessed_correctly:
        # Computer makes a guess
        guess = random.randint(low, high)
        attempts += 1

        # Get user feedback
        feedback = input(f"Is {guess} your number? (H)igh, (L)ow, or (C)orrect? ").upper()

        # Process feedback
        if feedback == 'H':
            high = guess - 1
            print("Okay, I'll guess lower next time!")
        elif feedback == 'L':
            low = guess + 1
            print("Okay, I'll guess higher next time!")
        elif feedback == 'C':
            guessed_correctly = True
            print(f"\nüéâ Hooray! I guessed your number in {attempts} attempts!")
        else:
            print("Please enter H, L, or C.")
            attempts -= 1  # Don't count invalid feedback as an attempt

        # Check if user is cheating
        if low > high:
            print("\nü§® Hey! You must have given me wrong feedback. There are no numbers left to guess!")
            return

# Start the game
computer_guesses_number()
```

```python
import random

def get_user_choice():
    """Get and validate user input"""
    while True:
        user_choice = input("Choose rock, paper, or scissors: ").lower()
        if user_choice in ['rock', 'paper', 'scissors']:
            return user_choice
        print("Invalid choice. Please try again.")

def get_computer_choice():
    """Generate computer's choice randomly"""
    return random.choice(['rock', 'paper', 'scissors'])

def determine_winner(user, computer):
    """Determine the winner of the round"""
    if user == computer:
        return 'tie'
    elif (user == 'rock' and computer == 'scissors') or \
         (user == 'paper' and computer == 'rock') or \
         (user == 'scissors' and computer == 'paper'):
        return 'user'
    else:
        return 'computer'

def play_game():
    """Main game function"""
    print("Welcome to Rock, Paper, Scissors!")
    print("First to win 3 rounds wins the game!\n")

    user_score = 0
    computer_score = 0

    while user_score < 3 and computer_score < 3:
        print(f"\nScore: You {user_score} - Computer {computer_score}")

        user_choice = get_user_choice()
        computer_choice = get_computer_choice()

        print(f"\nYou chose: {user_choice}")
        print(f"Computer chose: {computer_choice}")

        result = determine_winner(user_choice, computer_choice)

        if result == 'tie':
            print("It's a tie!")
        elif result == 'user':
            user_score += 1
            print(f"You win! {user_choice} beats {computer_choice}")
        else:
            computer_score += 1
            print(f"You lose! {computer_choice} beats {user_choice}")

    print("\n=== Final Score ===")
    print(f"You {user_score} - Computer {computer_score}")
    if user_score > computer_score:
        print("üéâ Congratulations! You won the game!")
    else:
        print("üò¢ The computer won the game. Better luck next time!")

    play_again = input("\nWould you like to play again? (yes/no): ").lower()
    if play_again == 'yes':
        play_game()
    else:
        print("Thanks for playing!")

# Start the game
play_game()
```

```python
import random
import string

def hangman():
    # Game setup
    word_list = ["python", "javascript", "hangman", "programming", "computer",
                 "keyboard", "developer", "algorithm", "function", "variable"]
    secret_word = random.choice(word_list).lower()
    letters_guessed = set()
    alphabet = set(string.ascii_lowercase)
    lives = 6
    hangman_stages = [
        """
           -----
           |   |
               |
               |
               |
               |
        """,
        """
           -----
           |   |
           O   |
               |
               |
               |
        """,
        """
           -----
           |   |
           O   |
           |   |
               |
               |
        """,
        """
           -----
           |   |
           O   |
          /|   |
               |
               |
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
               |
               |
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          /    |
               |
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          / \\  |
               |
        """
    ]

    print("Welcome to Hangman!")
    print("Guess the word one letter at a time. You have 6 lives.\n")

    while lives > 0:
        # Display current game state
        display_word = [letter if letter in letters_guessed else '_' for letter in secret_word]
        print(" ".join(display_word))
        print(hangman_stages[6 - lives])
        print(f"Lives left: {lives}")
        print(f"Letters used: {' '.join(sorted(letters_guessed))}\n")

        # Get user input
        while True:
            guess = input("Guess a letter: ").lower()
            if len(guess) != 1:
                print("Please enter a single letter.")
            elif guess not in alphabet:
                print("Please enter a valid letter.")
            elif guess in letters_guessed:
                print("You've already guessed that letter.")
            else:
                break

        letters_guessed.add(guess)

        # Check if guess is correct
        if guess in secret_word:
            print("\nGood guess!")
        else:
            lives -= 1
            print(f"\nWrong! The letter '{guess}' is not in the word.")
            print(f"You have {lives} lives remaining.")

        # Check if player has won
        if all(letter in letters_guessed for letter in secret_word):
            print(f"\nCongratulations! You won! The word was '{secret_word}'.")
            return

    # Player lost
    print(hangman_stages[6])
    print(f"\nGame over! You lost. The word was '{secret_word}'.")

# Start the game
hangman()
```

```python
import time

def countdown_timer():
    """Main function for the countdown timer"""
    print("=== PYTHON COUNTDOWN TIMER ===")

    # Get user input for timer duration
    while True:
        try:
            hours = int(input("Enter hours: ") or 0)
            minutes = int(input("Enter minutes: ") or 0)
            seconds = int(input("Enter seconds: ") or 0)

            if hours < 0 or minutes < 0 or seconds < 0:
                print("Please enter positive numbers only.")
                continue

            total_seconds = hours * 3600 + minutes * 60 + seconds
            if total_seconds <= 0:
                print("Please enter a duration greater than 0.")
                continue

            break
        except ValueError:
            print("Please enter valid numbers only.")

    # Countdown loop
    while total_seconds > 0:
        # Calculate time components
        hours = total_seconds // 3600
        remaining_seconds = total_seconds % 3600
        minutes = remaining_seconds // 60
        seconds = remaining_seconds % 60

        # Format the time display
        timer_display = f"{hours:02d}:{minutes:02d}:{seconds:02d}"
        print(timer_display, end="\r")  # \r returns to start of line

        # Wait one second
        time.sleep(1)

        # Decrement total seconds
        total_seconds -= 1

    # Timer complete
    print("\nTime's up!")
    for _ in range(3):  # Flash the message 3 times
        print("=== TIME'S UP ===")
        time.sleep(0.5)
        print("\033[F", end="")  # Move cursor up one line
        time.sleep(0.5)

# Start the timer
countdown_timer()
```

```python
import random
import string

def generate_password(length):
    """Generate a single random password with given length"""
    # Define character sets
    lowercase = string.ascii_lowercase
    uppercase = string.ascii_uppercase
    digits = string.digits
    symbols = string.punctuation

    # Combine all characters
    all_chars = lowercase + uppercase + digits + symbols

    # Ensure at least one character from each set
    password = [
        random.choice(lowercase),
        random.choice(uppercase),
        random.choice(digits),
        random.choice(symbols)
    ]

    # Fill the rest randomly
    for _ in range(length - 4):
        password.append(random.choice(all_chars))

    # Shuffle to avoid predictable patterns
    random.shuffle(password)

    # Convert list to string
    return ''.join(password)

def main():
    print("=== RANDOM PASSWORD GENERATOR ===")

    # Get user input
    while True:
        try:
            num_passwords = int(input("How many passwords do you want? "))
            length = int(input("Enter password length (minimum 8): "))

            if length < 8:
                print("Password must be at least 8 characters long.")
                continue
            if num_passwords < 1:
                print("Please enter at least 1 password.")
                continue
            break
        except ValueError:
            print("Please enter a valid number.")

    # Generate and print passwords
    print("\nGenerated Passwords:")
    for i in range(num_passwords):
        password = generate_password(length)
        print(f"{i + 1}. {password}")

if __name__ == "__main__":
    main()
```

```python
!pip install qrcode[pil] opencv-python pyzbar pillow

import qrcode
from PIL import Image
import cv2
from pyzbar.pyzbar import decode
import os

def generate_qr_code(data, filename="qrcode.png", fill_color="black", back_color="white"):
    """
    Generate a QR code from the given data and save it as an image file.

    Args:
        data (str): The data to encode in the QR code
        filename (str): Output filename (default: "qrcode.png")
        fill_color (str): Color of QR code (default: "black")
        back_color (str): Background color (default: "white")
    """
    try:
        # Create QR code instance
        qr = qrcode.QRCode(
            version=1,
            error_correction=qrcode.constants.ERROR_CORRECT_L,
            box_size=10,
            border=4,
        )

        # Add data to the QR code
        qr.add_data(data)
        qr.make(fit=True)

        # Create an image from the QR code
        img = qr.make_image(fill_color=fill_color, back_color=back_color)

        # Save the image
        img.save(filename)
        print(f"QR code generated successfully and saved as {filename}")
        return True
    except Exception as e:
        print(f"Error generating QR code: {e}")
        return False

def read_qr_code(filename):
    """
    Read a QR code from an image file.

    Args:
        filename (str): Path to the image file containing QR code

    Returns:
        str: Decoded data from QR code or None if not found
    """
    try:
        # Check if file exists
        if not os.path.exists(filename):
            print(f"Error: File '{filename}' not found")
            return None

        # Read the image
        img = cv2.imread(filename)

        # Decode the QR code
        decoded_objects = decode(img)

        if decoded_objects:
            # Return the first decoded object
            return decoded_objects[0].data.decode("utf-8")
        else:
            print("No QR code found in the image")
            return None
    except Exception as e:
        print(f"Error reading QR code: {e}")
        return None

def main():
    print("=== QR Code Generator & Decoder ===")

    while True:
        print("\nOptions:")
        print("1. Generate QR Code")
        print("2. Read QR Code")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ")

        if choice == "1":
            # QR Code Generation
            data = input("Enter data to encode in QR code: ")
            filename = input("Enter output filename (default: qrcode.png): ") or "qrcode.png"
            generate_qr_code(data, filename)

            # Show the generated QR code
            try:
                img = Image.open(filename)
                img.show()
            except:
                print("Could not display the image")

        elif choice == "2":
            # QR Code Reading
            filename = input("Enter filename of QR code image: ")
            result = read_qr_code(filename)
            if result:
                print(f"\nDecoded data: {result}")

        elif choice == "3":
            print("Exiting program...")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
from IPython.display import clear_output

def print_board(board):
    """Display the Tic-Tac-Toe board"""
    clear_output()
    print("\n")
    for i, row in enumerate(board):
        print(" " + " | ".join(row) + " ")
        if i < 2:
            print("-----------")
    print("\n")

def check_winner(board):
    """Check if there's a winner or tie"""
    # Check rows and columns
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return board[0][i]

    # Check diagonals
    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]

    # Check for tie
    if all(cell != " " for row in board for cell in row):
        return "Tie"

    return None

def minimax(board, depth, is_maximizing):
    """Minimax algorithm implementation"""
    result = check_winner(board)
    if result is not None:
        if result == "X":
            return -10 + depth
        elif result == "O":
            return 10 - depth
        else:
            return 0

    if is_maximizing:
        best_score = -np.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == " ":
                    board[i][j] = "O"
                    score = minimax(board, depth + 1, False)
                    board[i][j] = " "
                    best_score = max(score, best_score)
        return best_score
    else:
        best_score = np.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == " ":
                    board[i][j] = "X"
                    score = minimax(board, depth + 1, True)
                    board[i][j] = " "
                    best_score = min(score, best_score)
        return best_score

def ai_move(board):
    """Determine AI's move using minimax"""
    best_score = -np.inf
    move = None
    for i in range(3):
        for j in range(3):
            if board[i][j] == " ":
                board[i][j] = "O"
                score = minimax(board, 0, False)
                board[i][j] = " "
                if score > best_score:
                    best_score = score
                    move = (i, j)
    return move

def play_game():
    """Main game loop"""
    board = [[" " for _ in range(3)] for _ in range(3)]
    current_player = "X"  # Human starts first

    while True:
        print_board(board)
        winner = check_winner(board)
        if winner:
            if winner == "Tie":
                print("It's a tie!")
            else:
                print(f"{winner} wins!")
            break

        if current_player == "X":
            # Human's turn
            while True:
                try:
                    row = int(input("Enter row (0-2): "))
                    col = int(input("Enter column (0-2): "))
                    if board[row][col] == " ":
                        board[row][col] = "X"
                        break
                    else:
                        print("Cell already occupied!")
                except (ValueError, IndexError):
                    print("Invalid input! Enter numbers 0-2.")
        else:
            # AI's turn
            print("AI is thinking...")
            row, col = ai_move(board)
            board[row][col] = "O"

        current_player = "O" if current_player == "X" else "X"

# Start the game
print("Tic-Tac-Toe with Unbeatable AI")
print("You are X, AI is O")
play_game()
```

```python
import random
from IPython.display import clear_output

def computer_guess():
    """Main game function where computer guesses user's number"""
    print("üî¢ Guess the Number - Computer Edition üî¢")
    print("\nThink of a number between 1-100, and I'll try to guess it!")
    print("After each guess, tell me if I'm (H)igh, (L)ow, or (C)orrect!\n")

    # Initialize game variables
    low = 1
    high = 100
    attempts = 0
    guessed_correctly = False

    while not guessed_correctly:
        # Computer makes a guess (using binary search for efficiency)
        guess = random.randint(low, high)
        attempts += 1

        # Get user feedback
        while True:
            feedback = input(f"Is {guess} your number? (H/L/C): ").upper()
            if feedback in ['H', 'L', 'C']:
                break
            print("Please enter H, L, or C")

        # Process feedback
        if feedback == 'H':
            high = guess - 1
            print("I'll guess lower next time!")
        elif feedback == 'L':
            low = guess + 1
            print("I'll guess higher next time!")
        elif feedback == 'C':
            guessed_correctly = True
            print(f"\nüéâ I guessed your number in {attempts} attempts!")

        # Check for cheating
        if low > high:
            print("\nü§® You must have given me wrong hints! No possible numbers left!")
            return

        # Add slight delay for realism
        if not guessed_correctly:
            print("Thinking...\n")
            # clear_output()  # Uncomment if you want to clear between guesses

# Start the game
computer_guess()

# Option to play again
while True:
    play_again = input("\nPlay again? (Y/N): ").upper()
    if play_again == 'Y':
        clear_output()
        computer_guess()
    elif play_again == 'N':
        print("Thanks for playing!")
        break
    else:
        print("Please enter Y or N")
```
